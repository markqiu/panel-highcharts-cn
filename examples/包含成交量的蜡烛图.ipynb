```python
import urllib, requests, json

system_proxies = urllib.request.getproxies()

if system_proxies and 'http' in system_proxies:
	r = requests.get('https://data.jianshukeji.com/stock/history/000001', proxies={"https": system_proxies['http']})
else:
	r = requests.get('https://data.jianshukeji.com/stock/history/000001')

data = json.loads(r.text)
if data['code'] != 1:
	print('读取股票数据失败！')
	raise Exception('读取股票数据失败！')
data = data['data']
# split the data set into ohlc and volume
ohlc = []
volume = []
dataLength = len(data)
groupingUnits = [
		['week', [1]],
		['month', [1, 2, 3, 4, 6]]
]
i = 0
for i in range(len(data)):
	ohlc.append([
		data[i][0], # the date
		data[i][1], # open
		data[i][2], # high
		data[i][3], # low
		data[i][4] # close
	])
	volume.append([
		data[i][0],  # the date
		data[i][5]  # the volume
	])
ohlc
volume
```
